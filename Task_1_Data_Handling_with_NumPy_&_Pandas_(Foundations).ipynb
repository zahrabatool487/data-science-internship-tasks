```python
# TASK 1: Data Handling with NumPy & Pandas
# Environment: Google Colab

#LOAD DATASET
import pandas as pd

#DATASET LINK
url = "https://gist.githubusercontent.com/mattkram/2af421edd49f9d3109b2e83e71152887/raw/iris.csv"
df = pd.read_csv(url)
df.head()


```

```python
# INSPECT DATASET

df.shape
df.info()
df.columns
df.head()
```

```python
# Data Cleaning & Preparation

# Check for Missing Values
df.isnull().sum()

# Check for Duplicate Records
df.duplicated().sum()

# Remove duplicate records
df = df.drop_duplicates()
```

```python
# NumPy Array Operations

# Convert numeric columns to NumPy array:

numeric_data = df.iloc[:, :-1].to_numpy()
```

```python
# Basic Statistical Analysis (NumPy)

import numpy as np

np.mean(numeric_data)
np.median(numeric_data)
np.std(numeric_data)
```
